# Option 3: Recommendation


Task: Recommend item to the given customer id for a given date.
User Story: User should be able to provide a Customer ID and Date, and the program should be able to
recommend items to be purchased.

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Online Retail.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['Dates'] = pd.to_datetime(df['InvoiceDate']).dt.date
```

```python
df.head()
```

```python
df=df.drop(['InvoiceDate'], axis=1)
```

```python
df.head()
```

```python
df['Quantity']=abs(df['Quantity'])
```

```python
df=df[df['Quantity']>5]
```

```python
df.head()
```

```python
df['Dates']=df['Dates'].astype(str)
```

```python
pt = df.pivot_table(index=['CustomerID','Dates'],columns='Description',values='Quantity')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
pt.index[158]
```

```python
np.where(pt.index==(12346.0, '2011-01-18'))[0][0]
```

```python
CustomerID=12347.0
Dates='2010-12-07'
```

```python
np.where(pt.index==(CustomerID, Dates))[0][0]
```

```python
def recommend(CustomerID,Dates):
    # index fetch
    index = np.where(pt.index==(CustomerID, Dates))[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:20]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = df.loc[(df['CustomerID'] == pt.index[i[0]][0])&(df['Dates']==pt.index[i[0]][1]),'Description'].iloc[0]
        
        #item.extend(list(temp_df.drop_duplicates('Description')['Description'].values))       
        
        data.append(temp_df)
    
    return list(set(data))[0:4]
```

```python
recommend(18283.0,'2011-11-30')
```
